# Applied Pipeline

```python
# generate text
from transformers import pipeline

generator = pipeline("text-generation", model = 'distilgpt2')
generator("In this math course, we will tell you how to", max_length = 40, num_return_sequences = 2,)
```

```python
# fill a word
unmasker = pipeline('fill-mask')
data = unmasker("This <mask> will teach you math.", top_k = 10)
data[1]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'data' is the list of dictionaries from the previous cell
extracted_data = [{'score': item['score'], 'word': item['token_str']} for item in data]

df = pd.DataFrame(extracted_data)
fig = plt.figure(figsize = (10, 5))
sns.barplot(data = df, x = df.word, y = df.score)
fig.tight_layout()
plt.show()
```

```python
# sentiment analysis
classifier = pipeline('sentiment-analysis')
classifier(['I have been waiting for a HuggingFace course my whole life.',
            'I do not like this at all.'
            ])
```

```python
# Indentification
ner = pipeline('ner', grouped_entities = True)
ner('My name is Fikri and I work at Amazon in branch office at Manchester.')
```

# Behind Pipeline (PyTorch (pt) and TensorFlow (tf))

This page dedicated to unreveal what is behind the pipeline to deeply understand the complicated mechanism.

$$→ Tokenizer → Model → Postprocessing$$

$$Raw\space text → Input\space IDs → Logits → Predictions$$

$$This\space Course\space is\space Amazing → [101,2023,2607, 2003, 6429, 102] → [-4.3630, 4.68590] → POS:\space 99\%\space NEG: 1\%$$



```python
# raw text -> tokens -> special tokens -> input IDs
from transformers import AutoTokenizer
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

raw_inputs = [
    "I've been waiting for a HunggingFace course my whole life",
    "I hate this so much!",
]

inputs = tokenizer(raw_inputs, padding = True, truncation = True, return_tensors = 'pt') # Using PyTorch (pt)
print(inputs)
```

```python
# Input IDS -> Logits (raw, unnormalize scores that a model produces before applying an activation function)
from transformers import AutoModel

# use the same checkpoints
model = AutoModel.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits)
```

```python
import torch
prediction = torch.nn.functional.softmax(outputs.logits, dim = -1)
print(prediction*100)
```

```python
from transformers import AutoTokenizer
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

raw_inputs = [
    "I've been waiting for a HunggingFace course my whole life",
    "I hate this so much!",
]

inputs = tokenizer(raw_inputs, padding = True, truncation = True, return_tensors = 'tf') # Using TensorFlow (tf)
print(inputs)
```

```python
from transformers import TFAutoModel

model = TFAutoModel.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits)
```

```python
# predict with tensorflow
import tensorflow as tf
import numpy as np

prediction_1 = tf.math.softmax(outputs.logits, axis = -1)
print(np.round(prediction_1*100, 1))
```

```python
prediction_2 = tf.nn.softmax(outputs.logits, axis = -1)
print(np.round(prediction_2*100, 1))
```
